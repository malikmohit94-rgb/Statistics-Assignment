Q2 :What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?

Ans. The P-value is the probability of getting the observed results assuming the null hypothesis is true.

P-value (usually ≤ 0.05) means the observed data would be very unlikely if the null hypothesis were true.This gives evidence against the null hypothesis, so we reject it.

Ans . Z-test:

Used in situations involving large datasets or known population information.

Used in situations involving small samples or experiments where the population standard deviation is not known.

Q4  What is a confidence interval, and how does the margin of error influence
its width and interpretation?

Ans . A confidence interval is a range of values that is likely to contain the true population value (such as the true mean or proportion).

1. Margin of Error Affects the Width

Larger margin of error → Wider confidence interval

The estimate becomes less precise.

Happens with small sample sizes, high variability, or using a higher confidence level (like 99%).

Smaller margin of error → Narrower confidence interval

The estimate becomes more precise.

Happens with larger samples, low variability, or using a lower confidence level (like 90%).

Question 6: Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.

```python
import numpy as np
from scipy.stats import norm

data = np.array([52, 49, 50, 51, 53, 48, 50, 49, 52, 51])

mu_0 = data.mean()

sigma = data.std()

x_bar = np.mean(data)


n = len(data)


z = (x_bar - mu_0) / (sigma / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z)))

print("Sample Mean:", x_bar)
print("Z-value:", z)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis. There is no significant difference between the sample and population mean.")
```
